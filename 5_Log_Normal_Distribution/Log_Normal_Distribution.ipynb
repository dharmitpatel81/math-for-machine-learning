```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
df = pd.read_csv("income.csv", names=["income", "count"], skiprows=[0])
df.head()
```

```python
sns.set_theme(rc={'figure.figsize':(13, 10)})
g = sns.barplot(x='income', y='count', data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
```

```python
sns.set_theme(rc={'figure.figsize':(13, 10)})
g = sns.barplot(x='income', y='count', data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
g.set(xscale="log")
```
